```python
from pymongo import MongoClient
import os
import pandas as pd
```

```python
cwd = os.getcwd()
data_dir = os.path.join(cwd, '../data/')
```

```python
files = os.listdir(data_dir)

for csv in files:
    print(csv.split('.')[0])
```

```python
db_client = MongoClient('localhost', 27017)
```

```python
db = db_client['mydatabase']
```

```python
def insert_csv_to_mongo(csv_file, collection_name):
    df = pd.read_csv(data_dir + csv_file)
    collection = db[collection_name]
    data = df.to_dict(orient='records')
    for entry in data:
        collection.insert_one(entry)
    print(f"Data from {csv_file} inserted successfully into collection {collection_name}")
    
```

```python
def get_collection_data(collection_name):
    collection = db[collection_name]
    cursor = collection.find()
    data = list(cursor)
    return data
```
