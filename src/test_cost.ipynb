```python
from scipy.optimize import minimize_scalar

# Define your objective function
def objective_function(x):
    return (x) ** 2  # Example objective function, replace with your function

# Define the range for optimization
lower_bound = 2  # Lower bound of the range
upper_bound = 5  # Upper bound of the range

# Optimize for minimum within the specified range
result = minimize_scalar(objective_function, bounds=(lower_bound, upper_bound), method='bounded')

print("Optimal value within the range:", result.x)
print("Optimal function value:", result.fun)
```
