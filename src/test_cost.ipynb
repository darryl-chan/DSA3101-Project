```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function
def my_function(x):
    return np.sin(x)

# Generate x values
x = np.linspace(-2*np.pi, 2*np.pi, 100)

# Generate y values
y = my_function(x)

# Plot the graph
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Function Graph: y = sin(x)')
plt.grid(True)
plt.show()
```
